```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Linear_Regression.xlsx')
df
```

```python
from __future__ import division, print_function, unicode_literals
```

```python
X= df['Chieu cao (cm)']
y = df['Can nang (kg)']
plt.plot(X, y, 'ro')
plt.xlabel('Height')
plt.ylabel ('Weight')
```

=> Ta thấy rằng dữ liệu được sắp xếp gần như theo một đường thẳng => mô hình Linear có nhiều khả năng cho kết quả tốt   
- Ta có công thức:  cân nặng = w_1* (chieu cao) + w_0

- w_1 là hằng số cần tìm => bài toán Linear cần đi tìm các hệ tố tối ưu
- w_0 còn được gọi là bias
- mối quan hệ y= f(x) được gọi là mối quan hệ tuyến tính (linear)
- y là giá trị outcome 
- y hat là estimated value

Giả nghịch đảo của một ma trận A trong Python sẽ được tính bằng numpy.lialg.pinv(A), pinv là từ viết tắt cuarpseudo 
(Giả nghịch đảo (pseudo inverse) là trường hợp tổng quát của nghịch đảo khi ma trận không khả nghịch hoặc thậm chí không vuông).

```python
print(X.shape)   # a vecto with 15 entries
print(X.shape[0])  # retrieve the number of rows in array X
```

```python
X.shape
X_reshaped= X.values.reshape(-1,1)   # reshape into a vetor
X_reshaped

y_reshaped= y.values.reshape(-1,1)
```

```python
# Building Xbar
one = np.ones((X.shape[0], 1))   #Return a new array of given shape and type, filled with ones
Xbar= np.concatenate((one, X_reshaped), axis= 1)  # theo chieu cot
```

```python
# Calculating weights of the fitting line: 
A= np.dot(Xbar.T, Xbar)   # Xbar.T là chuyển vị của Xbar
b= np.dot(Xbar.T, y_reshaped)
w= np.dot(np.linalg.pinv(A), b)
print('w= ', w)     # intercept and slope
```

```python
w_0= w[0][0]
w_1= w[1][0]
x_0= np.linspace(145, 185, 2)
y_0= w_0+ w_1*x_0
```

```python
plt.plot(X, y, 'ro')
plt.plot(x_0, y_0)
plt.xlabel('Height')
plt.ylabel('Weight')
```
